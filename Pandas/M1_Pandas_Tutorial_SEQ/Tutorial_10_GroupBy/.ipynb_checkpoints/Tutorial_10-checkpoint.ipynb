# Topic : Group By

```python
import pandas as pd
df=pd.read_csv("data.csv")
df
```

### Find Max. temperature in each of the cities
### Find the average wind spped per city

```python
g= df.groupby("city")
g
```

```python
for city, city_df in g:
    print(city)
    print(city_df)
    print(city_df["day"].values)
```

```python
g.get_group("mumbai")
```
