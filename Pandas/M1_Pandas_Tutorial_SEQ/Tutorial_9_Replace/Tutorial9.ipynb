# Topics : Replace

```python
import pandas as pd
import numpy as np
df=pd.read_csv("data.csv")
df
```

```python
newdf=df.replace(-99999,np.NaN)
newdf
```

```python
newdf=df.replace([-99999,-8899],np.NaN)
newdf
```

```python
newdf=df.replace({"temperature": -99999,"windspeed":[-8899,-99999],"event": "No Event"},np.NaN)
newdf
```

```python
newdf=df.replace({"temperature": '[A-Za-z]',"windspeed":'[A-Za-z]',"event":'[R*]'},'**',regex=True)
newdf
```

```python
newdf=df.replace({-99999:np.NaN,"No Event":"Pradeep"})
newdf
```
