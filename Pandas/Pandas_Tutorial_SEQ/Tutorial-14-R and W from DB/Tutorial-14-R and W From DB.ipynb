# 3

```python
weather = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "Temp" : [32,45,30],
    "humidity":[80,60,70]
})
weather
```

```python
weather.rename(columns={
    "city":"Place",
    "Temp" :"Temp",
    "humidity" : "Humidity"
},inplace=True)
```

```python
weather
```

```python
weather.to_sql(
name="Weather",
con=engine,
index=False,
if_exists='append'
)
```
